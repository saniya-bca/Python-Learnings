```python
#Q.27]
# A B C D E
# J I H G F
# K L M N O
# T S R Q P
# U V W X Y

num = 0  # Counter for letters

for i in range(5):
    # Determine start letter for the row
    start = num
    if i % 2 == 0:  # Even row: forward
        for j in range(5):
            print(chr(65 + start + j), end=" ")
    else:           # Odd row: backward
        for j in range(4, -1, -1):
            print(chr(65 + start + j), end=" ")
    num += 5  # Increment counter for next row
    print()
```

```python
#Q.49
#        1
#       2  2
#     3     3
#    4       4
#   5         5

n = 5
for i in range(1, n + 1):
    # Print leading spaces
    for j in range(n - i):
        print(" ", end="")
    # Print first number
    print(i, end="")
    # Print spaces inside hollow part
    if i > 1:
        for j in range(2 * (i - 1) - 1):
            print(" ", end="")
        # Print last number
        print(i, end="")
    print()
```

```python
#Q.50
#     x       x
#      x     x
#       x   x
#        X X
#         x

n = 5
for i in range(1, n + 1):
    # Print leading spaces
    for j in range(i - 1):
        print(" ", end="")

    # Print first x
    print("x", end="")

    # Print inner spaces
    if i != n:
        for j in range(2*(n - i) - 1):
            print(" ", end="")
        # Print second x
        print("x", end="")

    print()
```
