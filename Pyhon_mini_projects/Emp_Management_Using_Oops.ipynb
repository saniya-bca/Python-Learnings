```python
class Employee():
  empcount=0
  incmt=1.10

  def __init__(self,first,last,salary):
    self.first=first
    self.last=last
    self.salary=salary
    self.email=first.lower()+"."+last.lower()+"gmail.com"
    Employee.empcount +=1
  def fullname(self):
    return f"{self.first.title()} {self.last.title()}"
  def increment(self):
    self.salary = int(self.salary * Employee.incmt)
```

```python
Emp1= Employee("saniya","shaikh",55000)
Emp2= Employee("siddhi","shetty",40000)
Emp3= Employee("swara","patil",30000)
```

```python
class Developer(Employee):
  def __init__(self,first,last,salary,prog_lang):
    super().__init__(first,last,salary)
    self.prog_lang = prog_lang
```

```python
class Manager(Employee):
  def __init__(self,first,last,salary,employees=None):
    super().__init__(first,last,salary)

    if(employees is None):
      self.employees=[]
    else:
      self.employees=employees
  def addemp(self,emp):
    if emp not in self.employees:
      self.employees.append(emp)
  def removeemp(self,emp):
    if emp not in self.employees:
      self.employees.remove(emp)
  def print_emps(self):
    for emp in self.employees:
      print('--->',emp.fullname())
```

```python
mng1=Manager('riya','verma',120000,[Emp1,Emp2,Emp3])
```

```python
mng1.print_emps()
```
